```python
import numpy as np
import pandas as pd
```

```python
# Set ipython's max row display
pd.set_option('display.max_row', 1000)

# Set iPython's max column width to 50
pd.set_option('display.max_columns', 50)
```

```python
file_name = 'sentiment_polarity/polarity_mturk_lsa_topics'
extension = '.csv'
df = pd.read_csv(file_name + extension, sep=',')
```

```python
df.head()
```

```python
df.describe()
```

```python
total_samples = len(df.id.unique())
min_samples = total_samples*0.1
print(min_samples)
```

```python
annotator_samples = df.groupby('annotator').id.count().reset_index()
annotator_samples.head()
```

```python
to_delete = annotator_samples[annotator_samples.id <= min_samples]
to_delete.head()
```

```python
# Filter Annotators
final_annotators = df[~df.annotator.isin(to_delete.annotator)].reset_index()
final_annotators = final_annotators.drop(columns=['index'])

# Sort Final Annotators
sorted_df = final_annotators.sort_values(by=['id'], ascending=True)
sorted_df = sorted_df.groupby('id').first()
sorted_df = sorted_df.drop(columns=['annotator','class'])
sorted_df.head()
```

```python
# Get Annotators Responses
annotators = final_annotators.pivot(index='id', columns='annotator', values='class')
annotators.head()
```

```python
sorted_df.to_csv('Databases/AMT/database_attr.csv', sep=',', index=False)
annotators.to_csv('Databases/AMT/database_labels.csv', sep=',', index=False)
```
